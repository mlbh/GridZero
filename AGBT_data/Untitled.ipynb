```python
import pandas as pd
import requests
url = "https://data.elexon.co.uk/bmrs/api/v1/generation/actual/per-type?from=2025-01-01&to=2025-12-31&settlementPeriodFrom=1&settlementPeriodTo=50&format=json"

response = requests.get(url)
json_data = response.json()

#df = pd.read_json(url)
df = pd.json_normalize(
    json_data['data'],
    record_path='data',
    meta=['startTime', 'settlementPeriod']
)
```

```python
df['StartTime'] = pd.to_datetime(df['startTime']).dt.tz_localize(None)
```

```python
df[(df['SettlementPeriod'] > 1) & (df['SettlementDate'] == '2025-01-01') & (df['PsrType'] == 'Biomass')]
```

```python
df_pivot = df.pivot_table(
    index='StartTime',
    columns='PsrType',
    values='Quantity',
    aggfunc='sum'
)
```

```python
df_pivot
```

```python
df_pivot['TotalOutput(MW)'] = df_pivot.sum(axis=1)
```
