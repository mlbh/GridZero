```python
import sys
import pandas as pd
sys.path.append('/Users/madeleine/code/mlbh/GridZero/backend')
```

```python
import importlib.util

spec = importlib.util.spec_from_file_location(
    "carbon_service",
    "/Users/madeleine/code/mlbh/GridZero/backend/app/services/carbon_service.py"
)
mod = importlib.util.module_from_spec(spec)  # ← fixed
spec.loader.exec_module(mod)
fetch_carbon_history = mod.fetch_carbon_history
```

```python
today = pd.Timestamp.now().floor("30min")
forecast_timestamps = pd.date_range(start=today, periods=48, freq="30min")

print("Fetching...")
carbon_history = fetch_carbon_history(forecast_timestamps)

print(f"Shape: {carbon_history.shape}")
print(f"NaNs: {carbon_history.isna().sum()}")
print(f"Sample:\n{carbon_history.head(5)}")
```

```python

lag_48    = forecast_timestamps - pd.Timedelta(hours=24)
lag_17520 = forecast_timestamps - pd.Timedelta(days=365)

print(f"Lag 48 NaNs:    {carbon_history.reindex(lag_48).isna().sum()}/48")
print(f"Lag 17520 NaNs: {carbon_history.reindex(lag_17520).isna().sum()}/48")
print(f"\nLag 48 sample:\n{carbon_history.reindex(lag_48).head(3)}")
print(f"\nLag 17520 sample:\n{carbon_history.reindex(lag_17520).head(3)}")
```
