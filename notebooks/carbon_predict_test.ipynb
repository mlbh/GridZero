```python
import importlib.util

spec = importlib.util.spec_from_file_location(
    "carbon_service",
    "/Users/madeleine/code/mlbh/GridZero/backend/app/services/carbon_service.py"
)
mod = importlib.util.module_from_spec(spec)  # ← fixed
spec.loader.exec_module(mod)
fetch_carbon_history = mod.fetch_carbon_history
```

```python
today = pd.Timestamp.now().floor("30min")
forecast_timestamps = pd.date_range(start=today, periods=48, freq="30min")

print("Fetching...")
carbon_history = fetch_carbon_history(forecast_timestamps)

print(f"Shape: {carbon_history.shape}")
print(f"NaNs: {carbon_history.isna().sum()}")
print(f"Sample:\n{carbon_history.head(5)}")
```

```python

lag_48    = forecast_timestamps - pd.Timedelta(hours=24)
lag_17520 = forecast_timestamps - pd.Timedelta(days=365)

print(f"Lag 48 NaNs:    {carbon_history.reindex(lag_48).isna().sum()}/48")
print(f"Lag 17520 NaNs: {carbon_history.reindex(lag_17520).isna().sum()}/48")
print(f"\nLag 48 sample:\n{carbon_history.reindex(lag_48).head(3)}")
print(f"\nLag 17520 sample:\n{carbon_history.reindex(lag_17520).head(3)}")
```

```python
import importlib.util
import pandas as pd
import numpy as np
import xgboost as xgb

spec = importlib.util.spec_from_file_location(
    "xgb_predictor",
    "/Users/madeleine/code/mlbh/GridZero/backend/app/models/xgb_predictor.py"
)
mod = importlib.util.module_from_spec(spec)
spec.loader.exec_module(mod)
XGBPredictor = mod.XGBPredictor
```

```python
model = xgb.XGBRegressor()
model.load_model('/Users/madeleine/code/mlbh/GridZero/notebooks/models_mlbh/xgb_full_06_1703.json')  # update this path
predictor = XGBPredictor(model)
```

```python
XGB_FEATURES = [
    "temperature_2m_c", "wind_speed_100m_ms", "wind_gusts_10m_ms",
    "cloud_cover_pct", "shortwave_radiation_wm2", "direct_radiation_wm2",
    "diffuse_radiation_wm2", "pressure_msl_hpa", "precipitation_mm",
    "biomass", "fossil_gas", "fossil_hard_coal", "hydro_pumped_storage",
    "hydro_run_of_river_and_poundage", "nuclear", "other", "solar",
    "wind_offshore", "wind_onshore", "totaloutput_mw",
    "hour_sin", "hour_cos", "dow_sin", "dow_cos", "doy_sin", "doy_cos",
    "carbon_lag_48", "carbon_lag_17520"
]
```

```python
fake_features = pd.DataFrame(
    np.random.rand(48, len(XGB_FEATURES)),
    columns=XGB_FEATURES
)
```

```python
predictions = predictor.predict(fake_features)

print(f"Number of predictions: {len(predictions)}")
print(f"First 5 values: {predictions[:5]}")
print(f"Min: {min(predictions):.1f}, Max: {max(predictions):.1f}")
```
