```python
import pandas as pd
import requests
```

```python
url_energy = 'https://data.elexon.co.uk/bmrs/api/v1/datasets/FUELINST?format=json'
url_carbon = 'https://api.carbonintensity.org.uk/intensity/2024-01-01/2024-01-02'
url_weather_f = 'https://api.open-meteo.com/v1/forecast?latitude=51.5&longitude=-0.1&hourly=wind_speed_10m,temperature_2m,shortwave_radiation'

# url fo r
url_weather_archive = 'https://archive-api.open-meteo.com/v1/archive?latitude=51.5&longitude=-0.1&start_date=2025-01-01&end_date=2025-12-31&hourly=temperature_2m,wind_speed_10m,shortwave_radiation&models=era5'
```

```python
# Carbon API data frame
carbon_data = requests.get(url_carbon).json()
carbon_data['data']

carbon_df = pd.json_normalize(carbon_data['data'])
carbon_df.head()
```

```python
# Energy data frame
energy_data = requests.get(url_energy).json()
energy_df = pd.json_normalize(energy_data['data'])
energy_df.head()
```

```python
# Weather Forecast API data frame for London
# Location: London; latitude: 51.49288; longitude: -0.102874756,
weather_f_data = requests.get(url_weather_f).json()

weather_f_df = pd.DataFrame(weather_f_data["hourly"])

print(weather_f_df.head())
```

```python
# Weather ERA5 Archive API data frame: 2025 year sample
# Date: 2025-01-01 to 2025-12-31
# Location: London; latitude: 51.5; longitude: 0
# Endpoint additon for historical data: &models=era5

weather_a_data = requests.get(url_weather_archive).json()

weather_a_df = pd.DataFrame(weather_a_data["hourly"])

print(weather_a_df.head())
```
