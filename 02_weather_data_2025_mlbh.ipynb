```python
from google.cloud import bigquery
import pandas as pd
import requests
```

```python
# Weather ERA5 Archive API data frame: 2025 year sample
# Date: 2025-01-01 to 2025-12-31
# Location: London; latitude: 51.5; longitude: 0
# Endpoint additon for historical data: &models=era5

url_weather_archive = 'https://archive-api.open-meteo.com/v1/archive?latitude=51.5&longitude=-0.1&start_date=2025-01-01&end_date=2025-12-31&hourly=temperature_2m,wind_speed_10m,shortwave_radiation&models=era5'

weather_a_data = requests.get(url_weather_archive).json()

weather_a_df = pd.DataFrame(weather_a_data["hourly"])

print(weather_a_df.head())
weather_a_df.shape
```

```python
PROJECT = 'gridzero-489711'
DATASET ='historical_weather'
TABLE = 'era5_weather_2025'

table = f'{PROJECT}.{DATASET}.{TABLE}'

client = bigquery.Client(project=PROJECT)

write_mode = 'WRITE_TRUNCATE'
job_config = bigquery.LoadJobConfig(write_disposition=write_mode)

job = client.load_table_from_dataframe(weather_a_df, table, job_config=job_config)
result = job.result()
```
