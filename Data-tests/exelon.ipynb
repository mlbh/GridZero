# Loading Data

```python
import pandas as pd
```

```python
url = 'https://data.elexon.co.uk/bmrs/api/v1/datasets/FUELINST?format=json'
```

```python
elexon_df = pd.read_json(url)
```

```python
elexon_df
```

```python
elexon_df['data']
```

```python
# expand the dictionaries inside the "data" column
df = pd.json_normalize(elexon_df["data"])
df
```
