```python
import pandas as pd
import requests
```

```python
url = 'https://api.open-meteo.com/v1/forecast?latitude=51.5&longitude=-0.1&hourly=wind_speed_10m,temperature_2m,shortwave_radiation'
```

## using pd.json_read

```python
data = pd.read_json(url)
data
```

```python
data['hourly']
```

```python
df = pd.DataFrame(data['hourly'])
df
```

```python
df["hourly"].apply(pd.Series).T
```

# Using request.get()

```python
response = requests.get(url).json()
response
```

```python
data = response['hourly']
data
```

```python
df = pd.DataFrame(data)
df
```
